Importamos las librerias 

```python
import cv2
import os
import numpy as np
from scipy.linalg import eig
```

Generamos las rutas donde esta el video y donde quedan las imagenes

```python
video_path = "Data/temporizador.mp4" #"Data/Temporizador.mp4"
output_folder = "Data/GrayscaleImages/" #"Data/GrayscaleImages"
output_folder_2="Data/representacion_imagenes/"
```

Definimos el el directrio donde se encuentra las imagenes 

```python
cap = cv2.VideoCapture(video_path)

# Inicializa un contador de fotogramas
frame_count = 0

# Inicializa una lista para almacenar las imágenes en forma de columnas
image_columns = []
```

Generamos el ciclo para procesar las imagenes, ajustar la resolución de las imagenes y generar la matriz 

```python
while frame_count <= 29:
    # Lee un fotograma del video
    ret, frame = cap.read()
    if not ret:
        break  # Sale del bucle cuando se han leído todos los fotogramas

    # Procesa el fotograma (en este caso, se convierte a escala de grises)
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Ajusta la resolución a 400x600
    gray_frame = cv2.resize(gray_frame, (600, 400))

    ###### CARGAR A LA MATRIZ  
    img_column = gray_frame.reshape(-1, 1)  # Convierte la imagen en una columna
    image_columns.append(img_column)

    # Guarda el fotograma como una imagen
    image_filename = output_folder+f'frame_{frame_count:02d}.png'  # Define el nombre del archivo
    cv2.imwrite(image_filename, gray_frame)  # Guarda el fotograma como una imagen

    frame_count += 1
```

```python
# Libera el objeto de captura y cierra la ventana de video
cap.release()
cv2.destroyAllWindows()
##############

# Convierte la lista de columnas de imágenes en una matriz H de tamaño 240000 × 30
H = np.hstack(image_columns)

# Calcula el promedio de las filas de H
mean_row = np.mean(H, axis=0, keepdims=True)

# Resta el promedio de cada fila de H para obtener la matriz A
A = H - mean_row
```

Hacemos la multriplicación A(Transpuesta)x A para generar los valores propios 

```python
# Ahora vamos a calcular la matriz A por su transpuesta (Matriz de covarianza) (nxn)
matriz = A.T@A
matriz
```

```python
print(matriz.shape)
```

Generamos los valores propios y vectores propios

```python
# Ahora calculamos sus valores prorios y sus vectores propios
eigenvalues, eigenvectors = eig(matriz)

print("Valores propios:")
print(eigenvalues)

print("Vectores propios:")
print(eigenvectors)
```

Generemos los vectores prodpios de la multiplicación de A x A tranpuesta

```python
Vetores_propios_calculados = A@eigenvectors
Vetores_propios_calculados
```

Ordenamos los valores propios de mayor a menor

```python
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = Vetores_propios_calculados[:, idx]
```

Genermos las imagenes y las guardamos en la carpeta "representación_imagenes"

```python
# Toma los dos vectores propios principales y conviértelos en imágenes
resolution = (400,600)
for i in range(30):
    # Convierte el vector propio en una matriz de imagen
    eigenvector_image = eigenvectors[:, i].reshape(resolution)

    # Normaliza la matriz de imagen para que los valores estén entre 0 y 255
    eigenvector_image = cv2.normalize(eigenvector_image, None, 0, 255, cv2.NORM_MINMAX)

    # Convierte la matriz de imagen en escala de grises
    eigenvector_image = cv2.convertScaleAbs(eigenvector_image)

    # Guarda la imagen
    image_filename = output_folder_2+f'eigenvector_{i}.png'
    cv2.imwrite(image_filename, eigenvector_image)
```
